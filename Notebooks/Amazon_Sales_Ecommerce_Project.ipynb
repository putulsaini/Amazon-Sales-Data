```python
# List of (column, symbol) pairs to be replaced
replacements = [
    ('discounted_price', '₹'),
    ('actual_price', '₹'),
    ('discount_percentage', '%'),
    # ('rating', '.'),  # Uncomment if needed
    ('rating_count', ',')
]

# Function to replace symbols in multiple columns
def replace_symbols(df, replacements):
    for col, symbol in replacements:
        df[col] = df[col].str.replace(symbol, "", regex=False)
    return df

# Apply the replacements
df = replace_symbols(df, replacements)

# Display the updated DataFrame
print(df)
```

```python

# Assuming df is your original DataFrame

# Split the 'category' column into multiple columns
df_split = df['category'].str.split('|', expand=True)

# Rename the new columns dynamically based on the actual number of columns
df_split.columns = [f'Category_{i+1}' for i in range(df_split.shape[1])]

# Concatenate the split columns with the original DataFrame
df = pd.concat([df, df_split], axis=1)

# Now, when you display the DataFrame, it should show the changes along with the original columns
print(df.head(2))
```

```python
#df['rating_count'].fillna(df['rating_count'].mean(), inplace=True)  # Replace NaN with mean
# df['rating'].fillna(df['rating'].mean(), inplace=True)  # Replace NaN with mean

# Convert the 'rating_count' column to numeric, forcing errors to NaN (if any non-numeric values exist)
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

# Replace NaN with mean value
df['rating_count'].fillna(df['rating_count'].mean(), inplace=True)

print(df['rating_count'].head())
```

```python
# Convert the 'rating_count' column to numeric, forcing errors to NaN (if any non-numeric values exist)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Replace NaN with mean value
df['rating'].fillna(df['rating'].mean(), inplace=True)

print(df['rating'].head())
```
