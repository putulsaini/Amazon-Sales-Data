# Step 5: Handling Outliers

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(amazon_data.columns)
```

```python
# List of numerical columns you want to plot
columns_to_plot = ['discounted_price', 'actual_price', 'discount_percentage']

# Loop through the columns and plot boxplots
for column in columns_to_plot:
    sns.boxplot(x=amazon_data[column])
    plt.title(f"Boxplot of {column}")
    plt.show()
```

```python
amazon_data[column] = pd.to_numeric(amazon_data[column], errors='coerce')
```

```python
# Now fill missing values with the mean
amazon_data[column].fillna(0)
```

```python
print(amazon_data.dtypes)
```

```python
Q1 = amazon_data[column].quantile(0.25)
Q3 = amazon_data[column].quantile(0.75)
IQR = Q3 - Q1
lower_whisker = max(amazon_data[column].min(), Q1 - 1.5 * IQR)
upper_whisker = min(amazon_data[column].max(), Q3 + 1.5 * IQR)

print(f"{column} Whisker values:")
print(f"Lower whisker: {lower_whisker}")
print(f"Upper whisker: {upper_whisker}")
```

```python
# Removing outliers using IQR method
Q1 = amazon_data[column].quantile(0.25)
Q3 = amazon_data[column].quantile(0.75)
IQR = Q3 - Q1

# Defining the bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filtering out the outliers
data = amazon_data[(amazon_data[column] >= lower_bound) & (amazon_data[column] <= upper_bound)]
print("imported successfully")
```
